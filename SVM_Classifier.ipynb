Demonstrate the working of SVM classifier for a suitable data set

```python
import pandas as pd
from sklearn.svm import SVC
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
```

```python
df=pd.read_csv(r"datasets\survey lung cancer (1).csv")
df.replace('YES',1,inplace=True)
df.replace('NO',0,inplace=True)
df.replace('M',1,inplace=True)
df.replace('F',0,inplace=True)
df
```

```python
lc=df['LUNG_CANCER']
lc=lc.to_numpy()
lc=lc.astype('int')
x=df.drop(['LUNG_CANCER'],axis=1)
x=x.to_numpy()
```

```python
clf=SVC(kernel='linear')
clf.fit(x,lc)
pred=clf.predict(x)
```

```python
report=classification_report(lc,pred,target_names=['CANCER','NO CANCER'])
print(report)
```

```python
print('ACCURACY=',accuracy_score(lc,pred))
```

```python
tp,tn,fp,fn=confusion_matrix(lc,pred).ravel()
print('Sens=',tp/(tp+fn))
print('spec=',tn/(tn+fp))
```
